```python
x = torch.randint(0, 9000, (3, 100))
```

```python

for i in range(20):
    y = m(x)
    loss = F.cross_entropy(y.transpose(-1 ,-2), x)
    loss.backward()
    print(loss)

    for p in m.parameters():
        if p.grad != None:

            p.data -= 0.01 * p.grad
```
